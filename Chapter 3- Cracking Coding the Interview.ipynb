```python
class Stack:
    
    def __init__(self):
        self.lst=[]
        self.size=0
        
    def peek(self):
        if not self.lst:
            return None
        return self.lst[-1]
    
    def isEmpty(self):
        return len(self.lst)==0
    
    def pop(self):
        return self.lst.pop()
    
    def push(self,item):
        self.lst.append(item)
        self.size+=1
        
    def __bool__(self):
        return bool(self.lst)
    
    def __len__(self):
        return len(self.lst)

    def __str__(self):
        
        return "\n[\n"+ "\n".join([f"{i}" for i in self.lst[::-1]])+"\n]"
        
```

```python
stack= Stack()

print("Adding ....")
for i in range (5):
    stack.push(i+3)
    
 

print("len: ",len(stack))

print("Items:", stack)

print("Peek: ",stack.peek())

print("Removing ....")

for i in range (3):
    stack.pop()

print("len: ", len(stack))

print("Items:", stack)

print("Peek: ",stack.peek())
```

```python
class FixedMultiStack:
    
    
    def __init__(self, capacityOfEachStack, NumberOfStacks):
        
        self.stack= [None for i in range(capacityOfEachStack*NumberOfStacks)]
        self.capacity=capacityOfEachStack
        self.NumberOfStacks=NumberOfStacks
        self.indexes=[i*self.capacity for i in range(NumberOfStacks)]
        self.sizes=[0 for i in range(NumberOfStacks)]
      
    # stack numbers starts from 0
    def peek(self, stackNumber):
        index=self.IndexOfTop(stackNumber)
        return self.stack[index]
        
    
    def push(self, stackNumber, value):
        
        if self.isFull(stackNumber):
            return False
        
        index=self.IndexOfTop(stackNumber)+1
        self.stack[index]=value
        self.sizes[stackNumber]+=1
        
    def pop(self, stackNumber):
        
        if self.isEmpty(stackNumber):
            return False
        
        index=self.IndexOfTop(stackNumber)
        value=self.stack[index]
        self.stack[index]=None        
        self.sizes[stackNumber]-=1
        return value
    
    def isFull(self, stackNumber):
        if self.sizes[stackNumber]== self.capacity:
            return True
        return False
        
    def isEmpty(self, stackNumber):
        if self.sizes[stackNumber]==0:
            return True
        return False    
        
    
    def IndexOfTop(self, stackNumber):
        index=stackNumber*self.capacity+self.sizes[stackNumber]-1
        return index
    
        
    def printStack(self):
        print("Full List:", self.stack)
        
        print("Sizes:", self.sizes)
```

```python
class Stack_with_Min_B(Stack):
    
    def __init__(self):
        
        super().__init__()
        self.minStack = Stack()
        
    def push(self, item):
        
        if not self.minStack or item <= self.minStack.peek():
            self.minStack.push(item)
            
        super().push(item)
        
    
    def pop(self):
        item = super().pop()
        if item == self.minStack.peek():
            self.minStack.pop()
        return item
        
    
    
```

```python
def sortStack(stack):
    
    sorted_stack= Stack()
    
    while stack:
        item= stack.pop()
        
        #  move smaller items to another stack
        while sorted_stack and sorted_stack.peek() < item:
            stack.push(sorted_stack.pop())
         
        # put the item
        sorted_stack.push(item)
            
    return sorted_stack
```

```python
# Driver Code

st= Stack()

for i in [30,10,5,15,0,12]:
    st.push(i)
 
print(st)
print("\nSorting ...")

ans= sortStack(st)
print(ans)
```

```python
import time


class Node:
    def __init__(self, data, next_node=None):
        self.data = data
        self.next_node = next_node

    def __str__(self):
        return str(self.data)


class LinkedList:
    def __init__(self, head=None):
        self.head = head

    def insert(self, node):
        if self.head is None:
            self.head = node
            return
        current_node = self.head
        while current_node.next_node is not None:
            current_node = current_node.next_node
        current_node.next_node = node

    def pop_head(self):
        if self.head is not None:
            head_to_pop = self.head
            self.head = self.head.next_node
            return head_to_pop

        return None

    def size(self):
        current_node = self.head
        size = 0
        while current_node is not None:
            size += 1
            current_node = current_node.next_node
        return size


# Animal Definitions


class Animal:
    def __init__(self, name):
        self.time_admitted = time.time()
        self.name = name


class Cat(Animal):
    pass


class Dog(Animal):
    pass


class AnimalShelter(LinkedList):
    def enque(self, animal):
        animal_node = Node(animal)
        self.insert(animal_node)

    def dequeue_any(self):
        return super().pop_head().data.name

    def dequeue_cat(self):
        if self.head and isinstance(self.head.data, Cat):
            return super().pop_head().data.name
        
        if not self.head or not self.head.next_node:
            return "Not Exist"
        
        previous_node = self.head
        current_node = self.head.next_node
        while current_node is not None:
            if isinstance(current_node.data, Cat):
                previous_node.next_node = current_node.next_node
                return current_node.data.name
            previous_node = current_node
            current_node = current_node.next_node
        return None

    def dequeue_dog(self):
        if self.head and isinstance(self.head.data, Dog):
            return super().pop_head().data.name
        
        if not self.head or not self.head.next_node:
            return "Not Exist"
        
        previous_node = self.head
        current_node = self.head.next_node
        while current_node is not None:
            if isinstance(current_node.data, Dog):
                previous_node.next_node = current_node.next_node
                return current_node.data.name
            previous_node = current_node
            current_node = current_node.next_node
        return None
```
